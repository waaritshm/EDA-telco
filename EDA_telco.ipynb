```python
telco.groupby(['Churn','PhoneService','InternetService','SeniorCitizen']).agg(
    avg_month = ('MonthlyCharges','mean'),
    avg_tenure = ('tenure','mean'),
    total_user = ('customerID','count')
)
```

**Observation:**

From the table above it can be concluded:
- From user no churn:
  * It was obtained data that of users aged > 65 years, most of them were telephone service subscribers, also subscribed to internet services with DSL and fiber optic networks contributing the largest average monthly costs, namely for DSL of `66,069118` and optical fiber of `93,610046`,
  * for users < 65th who also subscribe to telephone and internet, the largest service is fiber optic in the first place and dsl in the next position, with the largest average service fee of `94.036113` for fiber optic and for DSL `66.069118`, the number of users for dsl of `1326` and fiber optic of `1362` too
  * who do not subscribe to the internet but subscribe to the telephone for `1360` people

- From user Churn
  * It was obtained data that from users aged > 65 years, most of them were telephone service customers, also subscribing to internet services with fiber optics contributed the largest average monthly fee, namely `88.484224` with the number of users as much as `393` people
  * users aged < 65 years who subscribe to telephone and internet use fiber optics more by `904` people and with an average monthly fee of `87.970962`

- If viewed from the subscription period, the user churn time is shorter than the user no chur, it can also be seen from the number of user churn on the optical fiber, it can be concluded that the optical fiber is having problems or something that interferes so that there are many users who churn


```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 6))

## Are churned customers likely to get charged more?
plt.subplot(1,3,1)
sns.boxplot(telco['Churn'], telco['MonthlyCharges'])
plt.title('MonthlyCharges vs Churn')

## When do customers churn?
plt.subplot(1,3,2)
sns.boxplot(telco['Churn'], telco['tenure'])
plt.title('Tenure vs Churn')

## Are senior citizen more likely to churn?
plt.subplot(1,3,3)
counts = (telco.groupby(['Churn'])['SeniorCitizen']
  .value_counts(normalize=True)
  .rename('percentage')
  .mul(100)
  .reset_index())

plot = sns.barplot(x="SeniorCitizen", y="percentage", hue="Churn", data=counts).set_title('SeniorCitizen vs Churn')
```

**Observation:**

Data from the graph above can be concluded that:
1. From the monthly charge and churn graph, it can be concluded that most customers who churn pay monthly fees that are greater than no churn users.
2. From the graph of tenure and churn, it can be concluded that users who churn have a shorter subscription period than users who do not churn
3. From the graph of senior citizens with churn, it can be concluded that customers who churn are more dominant than those aged > 65 years than those aged < 65 years.
